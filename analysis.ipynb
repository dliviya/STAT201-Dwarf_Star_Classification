```r
library(tidyverse)
library(ggrepel)
library(readxl)
library(broom)
library(infer)
library(gridExtra)
```

```r
# I changed the column names in my csv file to make them more manageable
# Temperature = T
# Luminosity = L
# Radius = R
# Absolute Magnitude = A_M
# Star Type = Type
# Star Color = Color
# Spectral Class = Spectral_class

stars <- read.csv("6-class.csv")

glimpse(stars)
```

```r
stars <- mutate(stars, Type = as_factor(Type))
```

```r
dwarfs <- stars %>%
filter(Type != 5) %>%
filter(Type != 4) %>%
filter(Type != 3)
```

```r
L_box <- ggplot(dwarfs, aes(x = Type, y = L)) + geom_boxplot()
T_box <- ggplot(dwarfs, aes(x = Type, y = T)) + geom_boxplot()
grid.arrange(L_box, T_box, ncol=2)
```

```r
type_0 <- filter(stars, Type == 0)
type_1 <- filter(stars, Type == 1)
type_2 <- filter(stars, Type == 2)
```

### Type 0 Distributions

```r
L_0 <- ggplot(type_0, aes(x = L)) + geom_histogram(bins = 20)
T_0 <- ggplot(type_0, aes(x = T)) + geom_histogram(bins = 20)
grid.arrange(L_0, T_0, ncol=2)
```

### Type 1 Distributions

```r
L_1 <- ggplot(type_1, aes(x = L)) + geom_histogram(bins = 20)
T_1 <- ggplot(type_1, aes(x = T)) + geom_histogram(bins = 20)
grid.arrange(L_1, T_1, ncol=2)
```

### Type 2 Distributions

```r
L_2 <- ggplot(type_2, aes(x = L)) + geom_histogram(bins = 20)
T_2 <- ggplot(type_2, aes(x = T)) + geom_histogram(bins = 20)
grid.arrange(L_2, T_2, ncol=2)
```
